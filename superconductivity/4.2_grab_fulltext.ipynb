```python
import requests, io, zipfile, pathlib, shutil
doi = "10.1103/PhysRevLett.55.2471"
# institutino white-listed, no API

url = f"https://harvest.aps.org/v2/journals/articles/{doi}"
headers = {
    "Accept": "application/zip",            # or application/json, application/xml, etc.
    #"User-Agent": "mouyang-fetch/0.1 (your@email)"
}
# somehow I can only download zip
extract_into = pathlib.Path("aps_article")
if extract_into.exists():
    shutil.rmtree(extract_into)

with requests.get(url, headers=headers, stream=True, timeout=60) as r:
    r.raise_for_status()          # 200 or bust
    zipped_bytes = io.BytesIO(r.content)

    with zipfile.ZipFile(zipped_bytes) as z:
        z.extractall(extract_into)
        print("Unpacked files:", z.namelist())
```

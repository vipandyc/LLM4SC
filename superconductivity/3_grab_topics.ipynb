Next step, from the theory papers extract key topics, using NLP.

For finding new topics, https://maartengr.github.io/BERTopic/getting_started/zeroshot/zeroshot.html#example

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/SC_related_cites_010.csv")
df = df.dropna()

from bertopic import BERTopic
from bertopic.representation import KeyBERTInspired

docs = df["abstract"]

# We define a number of topics that we know are in the documents
# This combination works: zeroshot_topic_list = ["ferromagnetic spin fluctuation", "anti-ferromagnetic spin fluctuation","BCS Cooper pair", "bipolaron"]

zeroshot_topic_list = ["ferromagnetic spin fluctuation", "anti-ferromagnetic spin fluctuation",\
                       "bipolaron", "charge density wave", "strong electronic coupling", "electron phonon coupling"]

# We fit our model using the zero-shot topics
# and we define a minimum similarity. For each document,
# if the similarity does not exceed that value, it will be used
# for clustering instead.
topic_model = BERTopic(
    embedding_model="thenlper/gte-small",
    min_topic_size=5,
    zeroshot_topic_list=zeroshot_topic_list,
    zeroshot_min_similarity=0.8,
    representation_model=KeyBERTInspired()
)
topics, _ = topic_model.fit_transform(docs)
```

```python
a = topic_model.get_topic_info()
a.to_csv("zeroshot_topics.csv", index=False)
```
